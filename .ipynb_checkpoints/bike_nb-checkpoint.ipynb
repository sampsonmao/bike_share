```python
import numpy as np
import pandas as pd
import seaborn as sns
```

## EDA

Practicing with uncleaned data, so convert encoded labels back to labels themselves:

```python
train = pd.read_csv("train.csv")
train.head()
```

```python
test = pd.read_csv("test.csv")
test.head()
```

```python
season_str = train.season.map({
    1: "spring",
    2: "summer",
    3: "fall",
    4: "winter"
})

holiday_str = train.holiday.map({
    0: "No",
    1: "Yes"
})

workingday_str = train.workingday.map({
    0: "No",
    1: "Yes"
})

weather_str = train.weather.map({
    1: "Clear, Few clouds, Partly cloudy, Partly cloudy",
    2: "Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist",
    3: "Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds",
    4: "Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog" 
})
```

```python
train.loc[:,"season"] = season_str
train.loc[:,"holiday"] = holiday_str
train.loc[:,"workingday"] = workingday_str
train.loc[:,"weather"] = weather_str
```

```python
import datetime as dt

train['datetime'] = pd.to_datetime(train['datetime'])
train['datetime']= train['datetime'].map(dt.datetime.toordinal)
train['datetime'] = train['datetime'] - min(train['datetime'])

test['datetime'] = pd.to_datetime(test['datetime'])
test['datetime']= test['datetime'].map(dt.datetime.toordinal)
test['datetime'] = test['datetime'] - min(test['datetime'])
```

## Modeling

The variables `casual` and `registered` are redundant, as the sum is `count`, so they should be removed. I'm only interested in the total count for this project. `atemp` is also a bit redundant; there is almost a 1:1 relationship with `temp`, except that there is an `atemp` that corresponds to many `temp` as seen in the pairplot. So, `atemp` will be removed. Do the same for the test set.

```python
train_clean = train.drop(['casual', 'registered', 'atemp'], axis=1)
test_clean = test.drop(['atemp'], axis=1)
```

sklearn requires all categorical variables to be dummy coded:

```python
train_dummy = pd.get_dummies(train_clean, columns=['season', 'holiday', 'workingday', 'weather'])
test_dummy = pd.get_dummies(test_clean, columns=['season', 'holiday', 'workingday', 'weather'])

X = train_dummy.drop(['count'], axis=1)
y = train_dummy['count']
```

The data has to be scaled or normalized so that the MLE-finding algorithm can converge easier. Here I am also splitting the training set so we can do model evaluation.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import TweedieRegressor

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
pipe_poisson = make_pipeline(StandardScaler(), TweedieRegressor(power=1, link='log'))
pipe_poisson.fit(X_train, y_train) 
pipe_poisson.score(X_test, y_test)
```

```python
from sklearn.linear_model import LinearRegression

pipe_linear = make_pipeline(StandardScaler(), LinearRegression())
pipe_linear.fit(X_train, y_train) 
pipe_linear.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

pipe_forest = make_pipeline(StandardScaler(), RandomForestRegressor())
pipe_forest.fit(X_train, y_train) 
pipe_forest.score(X_test, y_test)
```

```python
importances = pipe_forest.steps[1][1].feature_importances_
feature_names = X_train.columns
d_for_df = {
    "feature" : feature_names,
    "importance" : importances
}

importance_df = pd.DataFrame(d_for_df)
```

```python
importance_plot = sns.barplot(data=importance_df, x="feature", y="importance")
a = importance_plot.set_xticklabels(importance_plot.get_xticklabels(), 
                          rotation=90)
```
